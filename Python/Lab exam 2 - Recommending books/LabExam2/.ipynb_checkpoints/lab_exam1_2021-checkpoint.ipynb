### The similarity algorithm

If `user_a` enjoyed reading a number of books that `user_b` also enjoyed reading, then when `user_a` recommends another book that `user_b` has not read, chances are that `user_b` would like to read it too. On the other hand, if the two users always tend to disagree about books, then `user_b` is unlikely to read a book that `user_a` recommends. 

You can calculate how similar two users are by treating each of their ratings as a vector and calculating the dot product of these two vectors. The dot product is simply the sum of the products of each of the corresponding elements.

For example, suppose we have 3 books in our database and

* `user_a` rated them `[5, 3, -5]` (here 5, 3, and -5 is `user_a`'s rating for books 1, 2 and 3 respectively),
* `user_b` rated them `[1, 5, -3]`, 
* `user_c` rated them `[5, -3, 5]`, and 
* `user_d` rated them `[1, 3, 0]`. 

The similarity between `user_a` and `user_b` is calculated as: 

$$(5 \times 1) + (3 \times 5) + (-5 \times -3) = 5 + 15 + 15 = 35.$$

The similarity between `user_a` and `user_c` is: 

$$(5 \times 5) + (3 \times -3) + (-5 \times 5) = 25 - 9 - 25 = -9.$$

The similarity between `user_a` and `user_d` is 

$$(5 \times 1) + (3 \times 3) + (-5 \times 0) = 5 + 9 + 0 = 14.$$

We see that if two people like a book (rating it with a positive number), it increases their similarity; and if two people dislike a book (both giving it a negative number), it also increases their similarity.

Once you have calculated the pair-wise similarity between `user_a` and every other user, you can then identify whose ratings are most similar to `user_a`'s. In the above case, `user_a` is most similar to `user_b`, so we would recommend to `user_a` the top books from `user_b`'s list that `user_a` hasn't already read.

```python
class User:
    users = []
    books = []
    
    def __init__(self, name, ratings, similarities = {}):
        self.name = name
        self.ratings = ratings
        self.similarities = similarities
        
    def __repr__(self):
        return f"{self.name.title()}"
    
    # Reading the users with their ratings into list 'users'
    def read_ratings(filename):
        with open(filename) as f:
            lines = f.read().splitlines()

        for i in range(0, len(lines), 2):
            name = lines[i].strip()
            ratings = lines[i+1].strip().split(' ')

            # Saving users with name and a list or ratings(ints)
            user = User(name, list(map(int,ratings))) 
            User.users.append(user)
    
    # Reading the books into list 'books'
    def read_books(filename):
        with open(filename) as f:
            lines = f.read().splitlines()

        for line in lines:
            stripped_line = line.strip().split(',')
            author = stripped_line[0]
            title = stripped_line[1]

            book = Book(author, title)
            User.books.append(book)
            
    # A method for writing the new user to the file(ratings.txt)        
    def write_to_file(filename, user):
        with open(filename, "a") as f:
            f.write("\n")
            f.write(user.name)
            f.write("\n")
            f.write(' '.join(str(item) for item in user.ratings))
            
    def printing_rate_table():
        list_of_labels = ["-5 - Hated it!", "-3 - Didn't like it!", "0 - Haven't read it!",
                         "1 - OK!", "3 - Liked it!", "5 - Really liked it!"]
        print("Rate table: ")
        for label in list_of_labels:
            print(label)
            
    # If the user does not exist, the user has to rate 20% of the books
    # and then be added to the list of users
    def rate_books(name, filename):
        # Selecting 20% of the books on random
        random_books = random.choices(User.books, k = int((1/5)*len(User.books)))
        # Rates table
        User.printing_rate_table()
        rates = []
        for book in random_books:
            # Inputting rate until it is correct
            rate = int(input(f"Rate \"{book.title}\" by {book.author}: "))
            while rate != -5 and rate != 5 and rate != -3 and rate != 3 and rate != 0 and rate != 1:
                print("Incorrect rate! Rate again!")
                rate = int(input(f"Rate \"{book.title}\" by {book.author}: "))
                
            rates.append(rate)
        
        # Creates a list of ratings with values 0
        ratings = [0 for elt in range(len(User.books))]
        # Adds the rated movies to the ratings of the new user
        for i, book in enumerate(random_books):
            index = User.books.index(book)
            ratings[index] = rates[i]
        
        new_user = User(name,ratings)
        User.users.append(new_user)
        User.write_to_file(filename, new_user) # Writing user's name and ratings to ratings.txt
        return new_user

            
    # Get the user whose name is the given name
    def find_user_by_given_name(name):
        for user in User.users:
            if user.name == name:
                return user
    
    # Returns the dot product of the ratings of the object and another user
    def similarity_alg(self, user):
        product = 0
        for i in range(len(self.ratings)):
            product += self.ratings[i] * user.ratings[i]
        
        return product
    
    # Returns a sorted in descending order dictionary with users
    # and their similarity score compared to the object
    def find_similarities(self):
        flag = False
        for user in User.users:
            if user.name != self.name:
                product = self.similarity_alg(user)
                self.similarities[user] = product
        # Sorts the dictionary by similarity score in descending order and alphabetically if 
        # there are users with the same similarity score
        self.similarities = dict(sorted(self.similarities.items(),
                                            key = lambda item: (item[1], item[0].name.title()), reverse = True))    
        return list(self.similarities)
    
    # Printing out the given number of recommended books
    # according to the similarity score with recursion 
    # and then written to output.txt
    def recommended_books(self, recommendations, counter = 0, recommended_books = [], index = 0):
        if counter >= recommendations:
            return 
        else:
            # Writing the results in output.txt
            #f = open("output.txt", "a")
            user = self.find_similarities()[index]
            if len(recommended_books) == 0 and index == 0:   
                print(f"Recommending based on the similarity algorithm.")
#                 f.write("\n")
#                 f.write("+"*47 + "\n")
#                 f.write("Recommending based on the similarity algorithm.\n")
#                 f.write("+"*47 + "\n")
            #f.write(f"Books recommended by {user.name}:\n")
            print(f"Books recommended by {user.name}:")
            for i in range(len(user.ratings)):
                if counter < recommendations:
                    if (user.ratings[i] == 3 or user.ratings[i] == 5) and self.ratings[i] == 0:
                        book = User.books[i]
                        if book.title not in recommended_books:
                            print(f"\t{book.title}, written by {book.author}")
                            #f.write(f"\t{book.title}, written by {book.author}\n")
                            recommended_books.append(book.title)
                            counter += 1
            #f.close()
            return self.recommended_books(recommendations, counter, recommended_books, index + 1)          
            
class Book:
    def __init__(self, author, title):
        self.author = author
        self.title = title
        
    def __repr__(self):
        return f"{self.author} - {self.title}"

def input_data():
    User.read_ratings("ratings.txt") # Reading the usernames and ratings from ratings.txt
    User.read_books("books.txt") # Reading all the books from books.txt
    # Name check
    name = input("What is your name?: ")
    while len(name) < 1:
        print("The name is incorrect! Enter again!")
        name = input("What is your name?: ")
        
    # Recommendations check
    try:
        recommends = int(input(f"How many recommendations between 0 and {len(User.books)}  (The default value is 10): ")) or 10
        if recommends < 0 or recommends > len(User.books):
            raise ValueError
    except ValueError:
        return f"The recommendations must be between 0 and {len(User.books)}"
        
    # Entering new user
    try:
        user = User.find_user_by_given_name(name)
        if user == None:
            raise TypeError
    except TypeError:
        print(f"\nThe user {name} does not exist so you will have to rate {int((1/5)*len(User.books))} books, selected on random!")
        user = User.rate_books(name, "ratings.txt")
        
    user.recommended_books(recommends)
    print(f"\nSimilarity of other users with {name}:")
    for sim in user.similarities:
        print(f"\t{sim} : {user.similarities[sim]}")
    
    # Writing the similarities with the specified user
#     f = open("output.txt", "a")
#     f.write("\n")
#     f.write(f"Similarity of other users with {user.name}:\n")
#     for element in user.similarities:
#         f.write(f"\t{element}:{user.similarities[element]}\n")
#     f.close()
    
input_data()
```

### (c) Analysis of time complexity

(n) - number of users.
    The time complexity with respect to the number of users is O(n) regardless of the existence of the user. Following the flow of the program, we can see that the average time to find a user in the users 
    list is O(n/2). The next method a user appears in is the similarity algorithm but the time with respect to number of 
    users there is constant, hence O(1). Then, to find the similarities of all the users compared to self without self we have to iterate
    through the whole list => O(n). Following the sorting of the dictionary we have O(n-1). The last time where the 
    list of users appears is recommending the books. We get a single user according to an index which means
    constant time O(1) and to print the required number of books we will need to iterate at most through all of the users - 1
    => O(n-1). Adding all of the parts together, the time complexity according to the number of users results in:
        O(n/2) + O(1) + O(n) + O(n-1) + O(1) + O(n-1) = O(3n+n/2) = O(7n/2) = O(n). This implies that the program is executing 
    in linear time complexity.
    
(m) - number of books.
    The final time complexity with respect to the number of books is O(m) but different computations occur if the username
    is already in the users list or not. If the user is already present in the list, then the first occurance is in finding
    the similarity between self and another user. The program iterates through all of the ratings of both users. The number
    of ratings depends on the number of books that exist in the file, hence O(m). Then again when recommeding books, 
    ratings of users are iterated which implies O(m). In the end a book at a specific index is returned
    => constant time, O(1). Adding all of the parts together, the time complexity is O(m) + O(m) + O(1) = O(2m) = O(m).
    If the username is not present in the list of users, then there are some calculations beforehand. Firstly, 20% of all
    the books are chosen at random. Iterating them is time complexity of O(log m). Then the program finds a book at a 
    specific location which is O(1). Consequently, the aforementioned time complexity is calculated again and when added
    together O(log m) + O(1) + O(m) = O(m).
    This implies that the time complexity with respect to the number of books is O(m) => linear.
